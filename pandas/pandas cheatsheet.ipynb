```python
calories = {"day1": 420, "day2": 380, "day3": 390}
myval = pd.Series(calories)    # makes a disctionary into a key value pair
print(myval)
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
mydf = pd.DataFrame(data)   # to convert a 2d array into a dataframe   # 2d array to rows and columns
print(mydf)
```

```python
print(mydf.loc[0])  # to locate one row
```

```python
print(mydf.loc[[0, 1]])  # to locate more than one row
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
mydf2 = pd.DataFrame(data, index = ['day1', 'day2', 'day3'])  # to assign specific indices to a dataframe # named index
print(mydf2)
```

```python
print(mydf2.loc['day2']) # to locate a named index
```
